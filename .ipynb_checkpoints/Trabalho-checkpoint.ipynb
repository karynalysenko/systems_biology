#### 1 d) Evaluate if single gene deletions enhance the production of the compound. Rank the mutants obtained according to the compound production capacity and growth performance.

```python
from cobra.flux_analysis import (single_gene_deletion, double_gene_deletion)
```

```python
model = read_sbml_model("sth.xml")
for i in model.genes:
    with model:
        i.knock_out()
        tmp=model.optimize()
        print(tmp.fluxes) #producao do composto + taxa de crescimento
    #print(i)
print(len(model.genes))
```

```python
abc=single_gene_deletion(model,model.genes)
```

```python
print(abc.growth)
```

```python
for i in abc.growth:
    if i != (float('1.042253')):
        print(i)
```

```python
print(single_gene_deletion(model,model.genes).growth[0])
```
