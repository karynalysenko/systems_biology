#### 1 d) Evaluate if single gene deletions enhance the production of the compound. Rank the mutants obtained according to the compound production capacity and growth performance.

```python
model_knockout = read_sbml_model("sth.xml")
for i in model_knockout.genes:
    with model_knockout:
        i.knock_out()
        tmp=model_knockout.optimize()
        #print(tmp.fluxes, tmp.status) #producao do composto + taxa de crescimento
        print(model_knockout.optimize()["EX_diact_e"])
print(len(model_knockout.genes))
```
